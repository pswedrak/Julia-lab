```julia
using Gadfly
using DataFrames
using DifferentialEquations
using CSV
```

```julia
function LotkaVolterra(du, u, p, t)
    du[1] = p[1] * u[1] - p[2] * u[1] * u[2]
    du[2] = -p[3] * u[2] + p[4] * u[1] * u[2]
end
```

Experiment 1:

```julia
u = [50.0, 50.0]
p = [0.55, 0.028, 0.8, 0.024]
t = (0.0, 100.0)
```

```julia
prob = ODEProblem(LotkaVolterra,u,t,p)
```

```julia
sol = solve(prob, RK4(), dt = 1);
```

```julia
df1 = DataFrame(t = sol.t, x = [sol.u[i][1] for i=1:length(sol.u) ], y = [sol.u[i][2] for i=1:length(sol.u) ], experiment = "exp1");
```

```julia
CSV.write("dataframe1.csv", df1)
```

```julia
function drawPhase(alldata)
    layers = Vector{Gadfly.Layer}()
    colors = [colorant"magenta", colorant"green", colorant"yellow", colorant"blue", colorant"black"]
    exps = unique(alldata[:experiment])
    
    for i in 1:length(exps)
        exp_i = alldata[alldata[:experiment] .== exps[i], :]
        usedColor = color(colors[i % length(colors)])
        push!(layers, layer(exp_i, x  = "x", y = "y", Geom.line(), Theme(default_color = usedColor))...)
    end
    
    set_default_plot_size(25cm, 25cm)
    plot(layers, Guide.XLabel("Prey"), Guide.YLabel("Predators"), Guide.Title("Population"), 
    Guide.manual_color_key("Parameters", ["[0.55, 0.028, 0.8, 0.024]", "[0.55, 0.038, 0.4, 0.024]", 
            "[0.45, 0.028, 0.7, 0.024]", "[0.75, 0.028, 0.9, 0.014]"], 
            ["magenta", "green", "yellow", "blue"]),)
end
    
```

```julia
drawPhase(alldata)
```
